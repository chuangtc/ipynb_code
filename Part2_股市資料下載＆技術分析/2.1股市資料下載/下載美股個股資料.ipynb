## Part 0: Import packages

```python
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib inline
```

## Part 1: Download stock data from Yahoo Finance

```python
stock = yf.download("GOOG", "2019-09-15", "2019-09-25")
stock.head()
```

## Part 2: Save the data as csv file

```python
stock.to_csv("stock.csv")
```

## Part 3: Visualize the trend of the stock

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(stock["Open"], color = "red")
plt.title("Google open price")
plt.show();
```

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(stock["Open"], color = "red")
plt.plot(stock["Close"], color = "green")
plt.title("Google open price")
plt.legend()
plt.show();
```
