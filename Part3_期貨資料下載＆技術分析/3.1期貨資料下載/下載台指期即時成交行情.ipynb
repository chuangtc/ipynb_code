- 資料來源：[期交所 台指期 一般交易時段行情](https://mis.taifex.com.tw/futures/RegularSession/EquityIndices/FuturesDomestic/)

- 台股即時資料下載：[【Python 爬蟲】台股即時成交行情免費下載｜盤中即時分析遠離韭菜命運！](https://youtu.be/axpXUzqOo9k)

## 1.1: 下載台指期即時報價(一般交易行情)

```python
url = "https://mis.taifex.com.tw/futures/api/getQuoteList"
```

```python
payload = {"MarketType":"0",
           "SymbolType":"F",
           "KindID":"1",
           "CID":"TXF",
           "ExpireMonth":"",
           "RowSize":"全部",
           "PageNo":"",
           "SortColumn":"",
           "AscDesc":"A"}
```

```python
res = r.post(url, json = payload)
res
```

```python
data = res.json()
data
```

```python
df = pd.DataFrame(data['RtData']['QuoteList'])
df
```

```python
df = df[["DispCName", "Status", "CBidPrice1", "CBidSize1", "CAskPrice1", "CAskSize1", "CLastPrice", "CDiff", "CAmpRate", "CTotalVolume", "COpenPrice", "CHighPrice", "CLowPrice", "CRefPrice", "CTime"]]
df
```

```python
df.columns = ['商品', '狀態', '買進', '買量', '賣出', '賣量', '成交價', '漲跌', '振幅%', '成交量', '開盤', '最高', '最低', '參考價', '時間']
df
```

## 1.2: 把下載的資料另存成csv檔

```python
df.to_csv('futures_regular_trading.csv')
```
