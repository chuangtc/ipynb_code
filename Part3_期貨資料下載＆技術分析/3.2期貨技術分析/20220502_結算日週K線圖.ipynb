# Part 1: 下載大盤資料
- 股海小英雄youtube影片：[【Python股市爬蟲 第1集】超簡單免費下載個股資料｜必學Yahoo財經yfinance套件](https://youtu.be/ZIFmDbIb-ro)


### 1.1 下載大盤 - 每週資料 

```python
import yfinance as yf
df_week = yf.download(tickers = '^TWII', period = '1y', interval = '1wk')
df_week = df_week[1:-1]
df_week
```

### 1.2 下載大盤 - 每日資料 

```python
df_day = yf.download(tickers = '^TWII', period = '1y', interval = '1d')
df_day
```

### 1.3 計算從週三到週三的週K線資料

```python
from datetime import datetime
df_day['weekday'] = df_day.index.strftime('%A')
df_day
```

```python
idx = df_day.loc[(df_day['weekday'] == 'Wednesday')].index
idx
```

```python
import pandas as pd
df_wed = pd.DataFrame(columns = ['週別日期', '開盤價', '最高價', '最低價', '收盤價'])
df_wed
```

```python
df_wed['週別日期'] = idx[:-1]
df_wed
```

```python
df_wed = df_wed.set_index('週別日期')
df_wed
```

```python
for i in range(len(idx) - 1):
    df = df_day.loc[idx[i]: idx[i + 1]]
    df_wed.loc[idx[i]] = [df['Open'][0], max(df['High']), min(df['Low']), df['Close'][-1]]
df_wed
```

# Part 2: 畫出大盤週Ｋ線圖
- 股海小英雄youtube影片：[【Python股市分析 第1集】用K線預測大盤趨勢｜最好用Python套件分享](https://youtu.be/FX6YAyOdEaE)

### 2.1 畫出大盤週三到週三的週Ｋ線圖

```python
import plotly.graph_objects as go

fig = go.Figure(data = [go.Candlestick(x = df_wed.index,
                                       open = df_wed['開盤價'],
                                       high = df_wed['最高價'],
                                       low = df_wed['最低價'],
                                       close = df_wed['收盤價'],
                                       increasing_line_color = 'red',
                                       decreasing_line_color = 'green')])
fig.update_layout(title_text = '大盤（週三到週三）週Ｋ線圖')
fig.show()
```

### 2.2 畫出大盤週一到週五的週Ｋ線圖

```python
import plotly.graph_objects as go

fig = go.Figure(data = [go.Candlestick(x = df_week.index,
                                       open = df_week['Open'],
                                       high = df_week['High'],
                                       low = df_week['Low'],
                                       close = df_week['Close'],
                                       increasing_line_color = 'red',
                                       decreasing_line_color = 'green')])
fig.update_layout(title_text = '大盤（週ㄧ到週五）週Ｋ線圖')
fig.show()
```

### 2.3 畫出大盤週Ｋ線圖

```python
from plotly.subplots import make_subplots

fig = make_subplots(
    rows = 2,
    cols = 1,
    subplot_titles = ("大盤（週三到週三）週Ｋ線圖", "大盤（週ㄧ到週五）週Ｋ線圖")
)

fig.add_trace(go.Candlestick(x = df_wed.index,
                             open = df_wed['開盤價'],
                             high = df_wed['最高價'],
                             low = df_wed['最低價'],
                             close = df_wed['收盤價'],
                             increasing_line_color = 'red',
                             decreasing_line_color = 'green'),
             row = 1,
             col = 1)

fig.add_trace(go.Candlestick(x = df_week.index,
                             open = df_week['Open'],
                             high = df_week['High'],
                             low = df_week['Low'],
                             close = df_week['Close'],
                             increasing_line_color = 'red',
                             decreasing_line_color = 'green'),
             row = 2,
             col = 1)
fig.update_xaxes(rangeslider_visible = False)
fig.update_layout(title_text = '大盤週Ｋ線圖', width = 800, height = 600)
fig.show()
```
